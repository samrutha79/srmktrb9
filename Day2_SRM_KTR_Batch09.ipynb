```python
Strings:
    1. Def: Sequence of characters is known as string.
    2. An empty string occupies 49 bytes of data
    3. Declaration [Initialization]:
        1. Using single quotes
        2. Using double quotes
            Note: ABove two ways can be used to give input value in single line of data
        3. Using triple quotes
            Note: Triple quotes will be allowed to initialize string in multi line of input
        4. Using str() => Used to convert into string from other types
        5. Using input() => Used to initialize the string from user input
    4. Accessing of string content:
        1. Single character [Indexing]:
            Syntax:
                stringVarName[indexValue]
                1. indexValue starts from 0 and ends with lengthOfSrting-1
            Eg:
                s = "JEEVAN"
                    0  1  2  3  4  5
                    J  E  E  V  A  N
                2. Is negative index supported by python?
                    Yesssss, Python supports negative indexing: [indexValue range: -lengthOfString to -1]
                s = "JEEVAN"
                    -6  -5  -4  -3  -2  -1
                     J   E   E   V   A   N
                Note:
                    Positive Indexing: Search character from left to right
                    Negative Indexing: Search character from right to left
        2. Group of characters [Slicing]:
            Syntax:
                variableName[startIndex:endIndex:stepCount]
                    startIndex: From where to begin [By default: 0]
                    endIndex: Where to end [By default: lengthOfString-1]
                    stepCount: How many charaters need to jump [By default: 1]
                Note: Make sure startIndex value should be lessthan endIndex value, Otherwise empty string will be displayed.
                Overcome:
                    Place negativeValue in place of stepCount value, Then the data will be printed in reverse order
                Q: Is negative values are supports in slicing?
                    Yes, It Supports
    5. Can you change the content of string?
        Nooooooooooooo, Because of strings are immutable
    6. Operations:
            + : Concatenation
            * : Repeatation
            len() : Identify the length of string
            min() : Identify the character which is min
            max() : Identify the character which is max
            enumerate()
    7. String handling methods:
        1. use dir(str) or dir(stringClassObject) to know list of string handling methods
        2. use help(str.methodNameWithoutBrackets) to know the syntax of method
```

```python
#capitalize(): Used to prepare first character of the string into upper case and rest are lower case
s = "SRM University"
print(s)
print(s.capitalize())
print()

#casefold(): Used to convert every character into lower case [ASCII[English] and UNICODE[Non-English]]
print(s)
print(s.casefold())
print()

#lower(): Used to convert characters into lower case [Only in ASCII data i.e. English character]
print(s)
print(s.lower())
print()

#swapcase(): Used to convert upper into lower and lower into upper
print(s)
print(s.swapcase())
print()

#title(): Makes every word first charatcer into upper case and rest are in lower case
s = "make the first character have upper case and the rest lower"
print(s)
print(s.title())
print()

#upper(): Converts into upper case
print(s)
print(s.upper())
print()
```

```python
#center(): Prints data in center of the given length
s = "PYTHON"
print(s.center(20))
print(s.center(20,'~'))
print()

#ljust():
print(s.ljust(20))
print(s.ljust(20,'^'))
print()

#rjust():
print(s.rjust(20))
print(s.rjust(20,'-'))
print()

#zfill(): It can fill empty spaces with '0'
print(s.zfill(20))
```
