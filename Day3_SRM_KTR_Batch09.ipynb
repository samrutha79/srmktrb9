```python
Tuple:
    1. Def: Ordered collection of data elements of same type or different types, But it is immutable is known as tuple
    2. Initialization:
        1. using () => prepares empty tuple, but there is no use of it
        2. using variable = list of values separated by comma [,]
        3. using (v1,v2,.....,vN)
        4. uisng tuple()
            1. from input
            2. from otherCollections
        **Note: As like as list, The tuple will store duplicate data elements also
    3. Accessing of tuple data:
        1. Indexing
        2. Slicing
    4. Is it possible to change the content of tuple?
        Noooooooooo, You cant change the content of tuple, But you can change elements of tuple element
    5. All basic operation could be applied on tuple
    6. Tuple handling methods:
        1. dir(tuple)
        2. help(tuple.methodName) / help(tuple)
```

```python
List Vs Tuple:
    1. List is mutable but Tuple is immutable
    2. List occupies more memory [Buffer space] but tuple occupies less memory compared to list
    3. Usage:
        1. If there is modification required on received information then we can use List
        2. If there is no modification required on received information then we can use tuple
```

```python
import time as t

st = t.time()
li = list(range(10000000))
et = t.time()
print("List preparation time is:",et-st)

st = t.time()
tu = tuple(range(10000000))
et = t.time()
print("Tuple preparation time is:",et-st)
```

```python
st = t.time()
for x in li:
    pass
et = t.time()
print("List accessing time is:",et-st)

st = t.time()
for x in tu:
    pass
et = t.time()
print("Tuple accessing time is:",et-st)
```

```python
#Eg: Prepare a dictionary with freqnecy of characters in given paragrapgh
d = dict()
for char in input():
    if char not in d:
        d[char] = 1
    else:
        d[char] = d[char] + 1
print(d)
```

```python
#Eg: Prepare a dictionary with frequncy of words in given paragraph
d = dict()
for word in input("Enter paragraph:").split():
    if word not in d:
        d[word] = 1
    else:
        d[word] = d[word] + 1
print(d)
```

```python
#Eg: Print keys of the dictionary
cu = {'Ind':'INR', 'US':'Dollar', 'UK':'Pound', 'Eur':'Euro', 'Swiss':'Franc', 'Canada':'CAD'}
print(cu.keys())

for key in cu:
    print(key,end=" ")

print()

for key in cu.keys():
    print(key,end=" ")
```

```python
#Eg: Print values of the dictionary
cu = {'Ind':'INR', 'US':'Dollar', 'UK':'Pound', 'Eur':'Euro', 'Swiss':'Franc', 'Canada':'CAD'}
print(cu.values())

for key in cu:
    print(cu[key],end=" ")

print()

for key in cu.values():
    print(key,end=" ")
```
