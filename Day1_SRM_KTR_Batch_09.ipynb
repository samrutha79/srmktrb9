```python
Variable(s) in Python:
    1. Def: A named memory location is known as variable
    2. Declaration of variable:
        Syntax:
            VariableName = literalConstant
            VariableName => Identifier
        Eg:
            #Valid Variables
            value1 = 152
            name = "Jeevan Kumar"
            stationFrequency = 98.3
            
            #Invalid Variables
            1number = 150
            second Number = 852
            station Frequency = 92.7
        Rules to define name to the variable:
            1. Name should begin with alphabet [A-Z a-z] or Underscore [ _ ]
            2. Name should not contain any special symbols except underscore [ _ ]
            3. Name should not be a keyword [Reserved word of the language]
            4. Name can be prepared with alphabets, digits and underscore but remember point number 1
    3. Knowing about type, size and address of variables:
        1. Type:
            Syntax:
                type(VariableName)
            Eg:
                x = 100
                type(x)
        
        2. Size:
            Syntax:
                import sys
                sys.getsizeof(variableName)
            Eg:
                import sys
                x = 100
                sys.getsizeof(x)
                
        3. Address:
            Syntax:
                id(variable)
            Eg:
                x = 100
                id(x)
    4. Ways of initialization:
        1. Single value into single variable
            Syntax:
                varName = constantValue
            Eg:
                x = 100
                name = "SRM"
        2. Single value into multiple variables
            Syntax:
                var1 = var2 = var3 = var4 = .... varN = Value
            Eg:
                v1 = v2 = v3 = 100
                name1 = name2 = "University"
        3. Multiple values into single variable
            Syntax:
                varName = value1, value2, value3,...., valueN #Packing of data
            Eg:
                var = 1,2,3,4,5 => Prepared a new tuple of data
        4. Multiple values into multiple variables
            Syntax:
                var1, var2, var3,...., varN = v1, v2, v3, .... , vN
                **Note: Total number of variable count should be equal to total number of values count or vice versa, Otherwise Error will be produces
            Eg:
                va,v2,v3 = 14,25,36
```

```python
Input and Output operations:
    1. Input:
        1. To read input from user, we can use a method known as input()
        2. To prompt the user for any specific data we can send prompt message to the input()
            Syntax:
                input(promptMessage with in double quotes)
        3. By default input() return type is string. So, If you want any specific data need to converted into specific type. [Typecasting]
            Syntax:
                int(input()) => Used to convert input value into integer
                float(input()) => Used to convert input value into float
                complex(input()) => Used to convert input value into complex
                list(input())
        4. Types of reading input:
            1. Reading single value into single variable:
                Syntax:
                    var = destinationTypeMethod(input("Prompt message")) #prompt message is optional
                Eg:
                    x = int(input("Enter an Integer:"))
            2. Reading single values into multiple variable:
                Syntax:
                    v1 = v2 = v3 = ..... = vN = destinationTypeMethod(input("Prompt message"))
                Eg:
                    x = y = z = int(input())
            3. Reading multiple values into single variable:
                Syntax1: By using split() of string class, But every input is string only
                    var = input().split() #List of data has been prepared type of string always
                    Eg: x = input().split()
                
                Syntax2: By Using list comprehension
                    var = [destinationTypeMethod(variable) for variable in collection]
                        collection:
                            1. input().split()
                            2. List of strings
                            3. List of tuples
                            .....
                
                Syntax3: By using map method #Can be used over another methods and we can stop using iteration
                    var = list(map(desinationType,inputSource))
                    Eg:
                        var = list(map(int,input("Enter numbers:").split()))
            4. Reading multiple values into multiple variables:
                Syntax1: var1,var2,var3,...,vanN = input(),input(),input(),.....,input()
                Syntax2: var1,var2,var3,...,vanN = input().split()
                Syntax3: var1,var2,var3,...,vanN = [destinationTypeMethod(variable) for variable in collection]
                Syntax4: var1,var2,var3,...,vanN = map(desinationType,inputSource)
                
                **Note: List of count in the left and right should be match, Otherwise an error will be produced
    2. Output:
        1. To print an output value on the screen, we can use a method print()
        2. Syntax of print():
            help(print)
        3. By default every print(), will print output value on separate line, If you want to print side by side we can use end=" "
```

```python
Operators supported in python:
    1. Arithmetic
    2. Relational
    3. Logical
    4. Bitwise
    5. Assignment
    6. Identity
    7. Membership
```

```python
'''
    Arithmetic:
        +  : Addition
        -  : Subtraction
        *  : MUltiplication
        /  : Float division
        // : Floor Division [Int Division]
        %  : Modulas [Remainder]
        ** : Exponansion [Power]
'''
print(10+20)
print(20-10)
print(20*30)
print(8/3)
print(8//3)
print(11%5)
print(4**3)
```

```python
'''
    Relational: Used to check the relation between two values / operands. Return value will be boolean.
        <  : Lessthan
        <= : Lessthan or Equal
        >  : Greaterthan
        >= : Greaterthan or Equal
        == : Is Equal
        != : Is not equal
'''
print(10<20)
print(10>20)
print(10<=20)
print(20>=20)
print(10==20)
print(10!=20)

print(10<20<30>15)
```

```python
'''
    Logical: Used to check two or more expressions at a time
        and: Returns True, In case result of all expressions are True, Otherwise False
        or : Returns False, In case result of all expressions are False, Otherwise True
        not : Returns True if input is False, False if input is True
'''
print(10<20 and 10<30)
print(True and True and True and True and False)

print(10>20 or 10<20)
print(False or False or False)

print(not True)
print(not False)
```

```python
'''
    Bitwise: Used to perform operations based on binary value [1 and 0] of a given number
        &  : Bitwise and => Returns 1, in case both bits are equal to 1, Otherwise 0
            Eg: x = 10 => 0000 1010
                y = 15 => 0000 1111
                x & y  => 0000 1010 => Result = 10  
        |  : Bitwise or => Returns 0, in case both bits are equal to 0, Otherwise 1
            Eg: x = 10 => 0000 1010
                y = 15 => 0000 1111
                x | y  => 0000 1111 => Result = 15
        ^  : Exclusive or => Returns 0, in case both bits are same, Otherwise 1
            Eg: x = 10 => 0000 1010
                y = 15 => 0000 1111
                x ^ y  => 0000 0101 => Result = 5
                0101 => 2**3 2**2, 2**1, 2**0
        << : left shift => Used to shift bits towards left side, Automatically value will be multipled by 2 after every shift
            Eg: x = 10 => 0000 1010
                x << 3:
                    x << 1: 0001 0100 => 20
                    x << 1: 0010 1000 => 40
                    x << 1: 0101 0000 => 80
            Result: inputValue * 2 power shiftValue
            Eg: x = 10
            Q: x << 3
                x<<3 = 10 * 2**3
                     = 10 * 8 => 80
        >> : right shift => Used to shift bits towards right side, Automatically value will be divided by 2 after every shift
            Result: inputValue // 2 power shiftValue
            Eg:
                x = 100
                x >> 4  = 100 // 2**4
                        = 100 // 16
                        = 6
        ~ : Negation [1's complement] => Used to give the out value in terms of 1's complement data
            Result: -((+ve/-ve)+1)
            Eg:
                x = 10
                ~x = -(10+1) = -11
                x = -10
                ~x = -(-10+1) = -(-9) = 9
'''
print(10 & 15)
print(10 | 15)
print(10 ^ 15)
print(10 << 3)
print(100 >> 4)
print(~10,~-10)
```

```python
'''
    Assignment: Used to store right side value into left side variable
        x = 100
        
        Arithmetic Assignment
        Bitwise Assignmet
        
        x = 100
        x += 25 => x = x + 25 => 100 + 25
        x &= 25 => x = x & 25 => 100 & 25
'''
x = 100
x += 25
print(x)

x = 100
x &= 25
print(x)
```

```python
'''
    Identity: Used to compare two objects
        is: Returns true in case both objects are referring towards same memory location [id]
        is not: Returns true in case both objects are not referring towards same memory location [id]
'''
print(10 == 10)
print(10 == 10.0)
print(10 == 10.000000000)

print(type(10), id(10), type(10.0,), id(10.0), id(10.0000000000))

print(10 is 10)
print(10 is 10.0)
print(10 is not 10.0)
```
