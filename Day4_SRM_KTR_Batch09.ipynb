```python
Exception - Handling:
    1. Types of errors:
        1. SYntax errors caused mismatch of syntax / pattern [Can be corrected]
        2. Runtime error caused due to logical [Invalid operation] mistakes [Can be managed]
    2. COns of Runtime errors:
        1. The program will be terminated abnormally
        2. Dont get any of the result [Success / Fail]
    3. Can we rectify the run-time errors:
        Noooooooooo, But we can manage it
        Due to the handling procedure we can intimate to end user about the run-time error.
    4. Handling of exceptions:
        1. try: Used to place the statements which can cause an exception(s)
            Syntax:
                try:
                    Statements
        2. except: Used to prepare the handling logic / statements in case of exception happens
            Syntax:
                as => Keyword introduced in python 3.x version
                except TypeOfException as objectName:
                    Statements
        3. finally:
    5. Handling multiple exceptions under single try block statements.
        Ans: There is a provision to place multiple except blocks under single try block. But minimum one except block is required
    6. Importance of finally block:
        1. Used to place the statements to be executed in case exception got raised or not raised.
    7. Is nested exception are possible?
        Yes, Possible to place one exception with in another exception
```

```python
Methods / Functions in Python:
    1. Def: A block of statements will be identified by name which can be executed whenever we used to call it.
    2. Syntax:
        def methodName(listOfParameter(s)):
            Statements
            return
        
        def: Keyword used to define new method
        methodName: Name of the method [Identifier]
        listOfParameter(s): Need to be passed if required
        Statements: Actual logic of method
        return: Keyword used to return the value after completion of the task by the method
    3. Advantages:
        1. Division of complex in to small pieces
        2. Re-use
        3. Improving of readability
        4. To perform unit - testing
    4. Types of parameters:
        1. Required / positional arguments
        2. Keyword arguments
        3. Default arguments
        4. Variable - length arguments
    5. Types of return values:
        1. Return single valued
        2. Return multi valued
    6. Nested functions:
        1. It is possible to place one function with in another function
        2. Inner function should be called from the body of outer function
        3. Inner function cannot be called from outside of outerfunction, If you try to do so which shows an error
    7. Anonymous methods:
        1. These methods are not having name, defined by using lambda keyword 
        2. Syntax:
            var = lambda listOfParam(s) : returnValue
        3. Is it possible to return multiple value by using anonymous methods?
            Noooooooo, NOt possible, But we can combile multiple values together and send it
```

```python
#variable - length argument:
#Eg: Calculate avg of 1st[3S+2L], 2nd[4S+3L] and 4th[2S+1L] year students

def avgOfStu(roll,name,*marks):
    print("Roll: ",roll)
    print("Name: ",name)
    print("Average: ",sum(marks)/len(marks))
    print()
    return

avgOfStu(1227,"Kumar",88,78,98,95,96)#1st
avgOfStu(1226,"Jeevan Kumar",88,78,98,95,96,88,89)#2nd
avgOfStu(1227,"Kumar",88,78,98)#4th
```

```python
Modules in python:
    1. Def: A module is a python program, can be collected with attributes, methods, classes
    2. Types of modules:
        1. Pre-defined modules: Readymadely available modules
        2. User-defined modules: Need to be prepared based on requirement
    3. Pre-defined modules:
        1. Already available
        2. Know the way [Syntax] how to use before using of it
        3. How to know list modules available in your environment
            Syntax:
                help('modules')
    4. How to use / import module into your program:
        1. Syntax1: 
                import moduleName
            Eg: import math
            Use: Modulename.methodName()
            Eg: math.factorial(5)
        2. Syntax2:
                import moduleName as aliasName
            Eg: import math as m
            Use: aliasName.methodName()
            Eg: m.factorial(5)
        3. Syntax3:
                from moduleName import methodName(s)
            Eg: from math import gcd, trunc
            Use: methodName()
            Eg: gcd(45,7)
        4. Syntax4:
                from moduleName import * # * => All
            Eg: from math import *
            Use: methodName() 
            Eg:  gcd(78,9)
```
